```python
df_dum = pd.get_dummies(df_model)
```

```python
df_dum.head()
```

```python
X = df_dum.drop('avg_salary', axis= 'columns')
Y = df_dum['avg_salary']
```

```python
X.info()
```

```python
# train test split 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#lasso model

from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
```

```python
lm_l = Lasso(alpha=.13)
lm_l.fit(x_train,y_train)
np.mean(cross_val_score(lm_l,x_train,y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

```python
#RandomForestRegressor model

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

np.mean(cross_val_score(rf,x_train,y_train,scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start = 400, stop = 1200, num = 5)]# Number of trees in random forest
max_features = ['auto','sqrt'] # Number of features to consider at every split
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)] # Maximum number of levels in tree
# max_depth.append(None)
min_sample_split = [2, 5, 10, 15, 100] #Minimum number of samples required to split a node
min_sample_leaf = [1, 2, 5, 10, 15, 20] #Minimum number of samples required at each leaf node
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_sample_split,
               'min_samples_leaf': min_sample_leaf}

print(random_grid)
```

```python
#Random search of parameters, using 3 fold cross validation,
#search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 5, verbose=2, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.score(x_test,y_test)
```

```python
predictions = rf_random.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
```
